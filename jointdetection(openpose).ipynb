```python
for i in range(xx,yy):
  folder_path =  f"Test Data/Dataset20/{i}"
  for filename2 in os.listdir(folder_path):
    image_path = os.path.join(folder_path, filename2)
    # print(image_path)
    
    df = pd.read_csv(file_path,low_memory=False)
    


# Define the body parts dictionary
    BODY_PARTS = {
        "Nose": 0, "Neck": 1, "RShoulder": 2, "RElbow": 3, "RWrist": 4,
        "LShoulder": 5, "LElbow": 6, "LWrist": 7, "RHip": 8, "RKnee": 9,
        "RAnkle": 10, "LHip": 11, "LKnee": 12, "LAnkle": 13, "REye": 14,
        "LEye": 15, "REar": 16, "LEar": 17, "Background": 18
    }
    
    # Input size for the neural network
    inWidth = 368
    inHeight = 368
    input_file = image_path  # Change this to the path of your input video or image file
    
    # Load the pre-trained network
    net = cv.dnn.readNetFromTensorflow("openpose/graph_opt.pb")
    
    # Open the input file (can be a video or an image)
    try:
        cap = cv.VideoCapture(input_file)
    except:
        continue
    while cv.waitKey(1) < 0:
        hasFrame, frame = cap.read()
        
        if not hasFrame:
            break
        
        frameWidth = frame.shape[1]
        frameHeight = frame.shape[0]
    
        net.setInput(cv.dnn.blobFromImage(frame, 1.0, (inWidth, inHeight), (127.5, 127.5, 127.5), swapRB=True, crop=False))
        out = net.forward()
        out = out[:, :19, :, :]  # MobileNet output [1, 57, -1, -1], we only need the first 19 elements
        
        # Extract the key points
        points = []
        for i in range(len(BODY_PARTS)):
            heatMap = out[0, i, :, :]
            
            _, conf, _, point = cv.minMaxLoc(heatMap)
            x = (frameWidth * point[0]) / out.shape[3]
            y = (frameHeight * point[1]) / out.shape[2]
            
            # Add the point to the list if the confidence is above a threshold
            points.append((int(x), int(y)))
        
        # Print the coordinates of the detected points
        break 
    # cc = 0
    # for i in points:
    #     if(i == (-1,-1)):
    #         cc+=1
    # if(cc>=9):
    #     continue
    new_row = {'ImgPath': image_path,
            "cx00" : points[0][0],"cy00" : points[0][1],
            "cx01" : points[1][0],"cy01" : points[1][1],
            "cx02" : points[2][0],"cy02" : points[2][1],
            "cx03" : points[3][0],"cy03" : points[3][1],
            "cx04" : points[4][0],"cy04" : points[4][1],
            "cx05" : points[5][0],"cy05" : points[5][1],
            "cx06" : points[6][0],"cy06" : points[6][1],
            "cx07" : points[7][0],"cy07" : points[7][1],
            "cx08" : points[8][0],"cy08" : points[8][1],
            "cx09" : points[9][0],"cy09" : points[9][1],
            "cx10" : points[10][0],"cy10" : points[10][1],
            "cx11" : points[11][0],"cy11" : points[11][1],
            "cx12" : points[12][0],"cy12" : points[12][1],
            "cx13" : points[13][0],"cy13" : points[13][1],
            "cx14" : points[14][0],"cy14" : points[14][1],
            "cx15" : points[15][0],"cy15" : points[15][1],
            "cx16" : points[16][0],"cy16" : points[16][1],
            "cx17" : points[17][0],"cy17" : points[17][1],
            "cx18" : points[18][0],"cy18" : points[18][1],
            'label' : image_path.split("/")[-2]}
    df = df.append(new_row, ignore_index=True)
    df.to_csv(file_path,index=False)
print(" done extracting")
```
